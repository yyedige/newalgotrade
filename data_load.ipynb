```python
import yfinance as yf
import pandas as pd
import os
```

```python
data = yf.download(tickers="AAPL", start='2000-01-01')
data
```

```python
out_dir = "data/raw"
os.makedirs(out_dir, exist_ok=True)
data.to_csv(os.path.join(out_dir, "data.csv"))
print(f"Saved {os.path.join(out_dir, 'data.csv')}")
```
