```python
import yfinance as yf
import pandas as pd
import os
```

```python
data = yf.download(tickers="NVDA", start='2000-01-01')
data
```

```python
out_dir = "data/raw"
os.makedirs(out_dir, exist_ok=True)
data.to_csv(os.path.join(out_dir, "NVDA.csv"))
print(f"Saved NVDA data to {os.path.join(out_dir, 'NVDA.csv')}")
```
